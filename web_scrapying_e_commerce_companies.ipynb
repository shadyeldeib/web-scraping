```python
# prompt: print more than num_results=10

def google_search(query, api_key, cse_id, num_results=10):
  """
  Performs a Google Custom Search API request.

  Args:
    query: The search query string.
    api_key: Your Google Custom Search API key.
    cse_id: Your Google Custom Search Engine ID.
    num_results: The maximum number of search results to return. Note that the API
                 returns a maximum of 10 results per request.

  Returns:
    A list of dictionaries containing the search results,
    or an empty list if the request failed or no results were found.
  """
  all_results = []
  start_index = 1  # The starting index for the results
  max_results_per_request = 10

  while len(all_results) < num_results and start_index <= 100: # Max start index is 100 for free API
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        'key': api_key,
        'cx': cse_id,
        'q': query,
        'num': min(num_results - len(all_results), max_results_per_request), # Request up to 10 results or remaining needed
        'start': start_index
    }

    try:
      response = requests.get(url, params=params)
      response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
      search_data = response.json()

      if 'items' in search_data:
        all_results.extend(search_data['items'])
        start_index += len(search_data['items']) # Increment start index for the next page
      else:
        # No more results for this query
        break

    except requests.exceptions.RequestException as e:
      print(f"Error during Google Search API request: {e}")
      break # Stop if there's an error

  return all_results

# Example Usage with more than 10 results:
search_query = "site:linkedin.com/in E-commerce Companies in Egypt "
num_results_to_get = 1000 # Specify the number of results you want
search_results = google_search(search_query, API_KEY, CSE_ID, num_results=num_results_to_get)


if search_results:
  print(f"Found {len(search_results)} search results for '{search_query}':")
  for i, item in enumerate(search_results):
    print(f"--- Result {i + 1} ---")
    print(f"Title: {item.get('title')}")
    print(f"Link: {item.get('link')}")
    print(f"Snippet: {item.get('snippet')}")
else:
  print("Could not perform the search or no results found.")

```

```python
import random
import nltk
from transformers import pipeline

# Download once
nltk.download('wordnet')
from nltk.corpus import wordnet

# Pre-defined base query
base_query = "E-commerce companies in Egypt"
site_prefix = "site:linkedin.com/in"

# Simple synonym generator using WordNet
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if "_" not in lemma.name():  # Skip multi-word terms
                synonyms.add(lemma.name().lower())
    return list(synonyms)

# Extract core keywords
keywords = ["e-commerce", "companies", "Egypt"]

# Manually correct synonyms for domain relevance
synonym_map = {
    "e-commerce": ["online stores", "digital retailers", "eCommerce firms"],
    "companies": ["businesses", "startups", "enterprises", "firms"],
    "Egypt": ["in Egypt", "Egyptian", "based in Egypt"]
}

# Generate variations
def generate_variations(keywords, max_variations=10):
    variations = set()
    while len(variations) < max_variations:
        terms = []
        for word in keywords:
            if word in synonym_map:
                terms.append(random.choice(synonym_map[word]))
            else:
                terms.append(word)
        variations.add(f"{site_prefix} {' '.join(terms)}")
    return list(variations)

# Run
query_variants = generate_variations(keywords)
for q in query_variants:
    print(q)
```

```python
import requests
import pandas as pd
import time  # For optional request delay

API_KEY = "***********"
CSE_ID = "*************"

queries = [
    "site:linkedin.com/in eCommerce firms enterprises based in Egypt",
    "site:linkedin.com/in digital retailers firms based in Egypt",
    "site:linkedin.com/in digital retailers businesses in Egypt",
    "site:linkedin.com/in eCommerce firms startups based in Egypt",
    "site:linkedin.com/in online stores enterprises based in Egypt",
    "site:linkedin.com/in online stores enterprises Egyptian",
    "site:linkedin.com/in digital retailers startups Egyptian",
    "site:linkedin.com/in eCommerce firms firms based in Egypt",
    "site:linkedin.com/in digital retailers startups based in Egypt",
    "site:linkedin.com/in online stores startups based in Egypt"
]

all_data = []

for q in queries:
    for start in range(1, 100, 10):  # Get up to 100 results per query
        params = {
            'key': API_KEY,
            'cx': CSE_ID,
            'q': q,
            'num': 10,
            'start': start
        }

        try:
            response = requests.get("https://www.googleapis.com/customsearch/v1", params=params)
            response.raise_for_status()
            results = response.json().get("items", [])

            for item in results:
                data = {
                    "title": item.get("title"),
                    "link": item.get("link"),
                    "snippet": item.get("snippet"),
                    "displayLink": item.get("displayLink"),
                    "query_used": q
                }
                all_data.append(data)

            time.sleep(10)  # Optional: respect rate limits

        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
            continue

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Export to CSV
df.to_csv("linkedin_search_results.csv", index=False)

print("✅ Exported search results to 'linkedin_search_results.csv'")
```
